```
from pyspark.sql.types import *
import  pyspark.sql.functions as F

# Create JSON Structure of payload (must know beforehand)
jsonSchema = (
               StructType([ 
                          StructField("EquipmentId", StringType(), True), 
                          StructField("TagId", StringType(), True), 
                          StructField("Value", StringType(), True), 
                          StructField("Type", StringType(), True), 
                          StructField("Units", StringType(), True), 
                          StructField("Timestamp", TimestampType(), True), 
                          StructField("Status", StringType(), True), 
                          StructField("UnixTimeMilliseconds",IntegerType(), True), 
                          StructField("UnixTimeSeconds",IntegerType(), True), 
                          StructField("UtcTimestamp", TimestampType(), True)
                          ])
             )

# Read stream cames in coded and add correct structure to the message
decoded_df = (
              corphistorianhub3
              .select( #Select from readStream object
                  F.from_json( 
                    F.col("body") # Only capture body field from eventhub schema
                      .cast("string"), jsonSchema
                              ) #Using structure in jsonSchema
                  .alias("payload")
                     ) #Add alias to all columns, creating a struct for the DataFrame
              )
#display(decoded_df) #Displaying Structured Streaming DataFrames will force the execution of Streaming Jobs
```

```
# Define a function to flatten out struct DataFrame
import pyspark.sql.functions as F
def flatten_df(nested_df):
    flat_cols = [c[0] for c in nested_df.dtypes if c[1][:6] != 'struct']
    nested_cols = [c[0] for c in nested_df.dtypes if c[1][:6] == 'struct']

    flat_df = nested_df.select(flat_cols +
                               [F.col(nc+'.'+c).alias(nc+'_'+c)
                                for nc in nested_cols
                                for c in nested_df.select(nc+'.*').columns])
    return flat_df
```

Create Structured Streaming Dataframe Object and TempView for SQL enrichment

Join Structured Streaming DataFrame with Static DataFrame and produce new Structured Streaming DataFrame Object

```
from pyspark.sql.types import IntegerType
def char_find(s):
  return s.rfind('_')
spark.udf.register("char_pos", char_find, IntegerType())
```

```
sqlContext.setConf("spark.sql.shuffle.partitions","8")
sqlContext.setConf("spark.sql.adaptive","enabled")
sqlContext.setConf("spark.sql.optimizer.dynamicPartitionPruning","enabled")
```

```
import pyspark.sql

#spark.sql.adaptive.enabled
#spark.sql.optimizer.dynamicPartitionPruning.enabled

df = spark.sql("""
with split as (
select
 			SUBSTRING(payload_TagId,char_pos(payload_TagId)+2,LENGTH(payload_TagId)) as UDC	
 		,	LEFT(LEFT(payload_TagId,CHARINDEX(':',payload_TagId)-1),CHARINDEX('.',payload_TagId)-1) as site
        ,   SUBSTRING(payload_TagId,CHARINDEX('.',payload_TagId)+1,LENGTH(payload_TagId) - CHARINDEX('.', payload_TagId) - CHARINDEX(':', REVERSE(payload_TagId))) as service
 		,	LENGTH(payload_TagId) - LENGTH(REPLACE(payload_TagId, ':', '')) as count
 		, 	*
from ch_stream
)

, ch_struct as (
select
	e.payload_UtcTimestamp,
	case 
		when count >= 1 then site
		when count = 0 then null
		else null
	end as site,
    case 
		when count >= 1 then service
		when count = 0 then null
		else null
	end as service,
	e.payload_EquipmentId,
	e.UDC,
	e.payload_Value,
    e.payload_Units,
    e.payload_Status
from split e 
)

, ch_rich as(
  select 
    *,
    LEFT(RIGHT(payload_EquipmentId,CHARINDEX('-',REVERSE(payload_EquipmentId))-1),CHARINDEX('_',payload_EquipmentId)-3) AS ASSET_CODE
  from ch_struct
  WHERE LEFT(payload_EquipmentId, 1) = 'M'

)

, ch_silver as(
  SELECT CA.*, A.ASSET_UID, A.ASSET_ID 
  FROM asset A
  INNER JOIN ch_rich CA ON A.ASSET_CODE = CA.ASSET_CODE
)
select * from ch_silver
""")
```
